```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from sklearn.calibration import calibration_curve
```

```python
### CLR
## Environmental Match

env_df = pd.read_csv('/data/tide/projects/ho_infxn_ml/results/model_results/20241210/environmental_colonization_pressure_with_pred.csv')

for run_name in env_df['run'].unique():

    run_df = env_df[env_df['run'] == run_name]

    true_pos, pred_pos = calibration_curve(run_df['group_binary'], run_df['pred'], n_bins=10)

    plt.figure()
    ax = plt.subplot(111)
    
    plt.plot(pred_pos,
             true_pos, 
             marker='o', 
             linewidth=1, 
             # label=f'{run_name} Conditional Logistic Regression',
            )

    plt.plot([0, 1], 
             [0, 1], 
             linestyle='--', 
             # label='Perfectly Calibrated',
            )

    plt.xlim(0,1)
    plt.ylim(0,1)
    
    plt.xlabel("Predicted Probability")
    plt.ylabel("True Probability")
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.title(f'{run_name} Calibration Plot')
    # plt.legend()
    plt.savefig(f'/PHShome/zw852/hospital_onset_personal_folder/Calibration_Plots/CLR/environmental_{run_name}_calibration.png')
```
