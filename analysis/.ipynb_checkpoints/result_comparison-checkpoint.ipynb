```python
import pandas as pd
import numpy as np
```

```python
old_coef = pd.read_csv('/data/tide/projects/ho_infxn_ml/results/model_results/20241210/clr_coefficients_removed_prior_pathogen.csv')
old_coef
```

```python
new_coef = pd.read_csv('/data/tide/projects/ho_infxn_ml/results/model_results/20250205/clr/clr_coefficients.csv')
new_coef.columns = ['match','target','variable','elix5_coef','elix5_lower_CI','elix5_upper_CI','elix5_SE_coef','elix5_pval','elix5_sig_flag']
new_coef
```

```python
combined_coef = old_coef.merge(new_coef,
                               on = ['match','target','variable'],
                               how = 'right')
combined_coef
```

```python
combined_coef.to_csv('/data/tide/projects/ho_infxn_ml/results/model_results/20241207/clr_coefficients_comparison.csv',
                    index = False)
```
